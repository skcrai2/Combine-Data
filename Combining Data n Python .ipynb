```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tn = pd.read_csv("data/TN.TXT", names=["state","sex","year","name","count"])
tn.head()
```

```python
ky = pd.read_csv("data/KY.TXT", names=["state","sex","year","name","count"])
ky.head()
```

```python
al = pd.read_csv("data/AL.TXT", names=["state","sex","year","name","count"])
al.head()
```

```python
ga = pd.read_csv("data/GA.TXT", names=["state","sex","year","name","count"])
ga.head()
```

```python
#2 Concatenate the four datasets

four_states = pd.concat([tn, ky, al, ga])
four_states
```

```python
#3 pivot table to answer question about number of males and females in each state with total for all four states

table = pd.pivot_table(four_states,index=["state"],values=["count"],columns=["sex"],aggfunc=[np.sum], fill_value=0, margins=True)
table
```

```python
#4 pivot table to get number of people born in each state and total

table2 = pd.pivot_table(four_states,index=["state"],values=["count"],aggfunc=[np.sum], fill_value=0, margins=True)
table2
```

```python
#5 pivot table for top 5 most popular names for any sex (assuming that question is referring to entire dataset)

table = pd.pivot_table(four_states,index=["name", 'sex'], values=["count"], aggfunc=[np.sum], fill_value=0)
report = table.sort_values([('sum','count')], 
               ascending=[False])
report.iloc[:5,:]
```

```python
#6 pivot table of Top 5 female names (assuming that question is referring to entire dataset)

report.query('sex == ["F"]').iloc[:5,:]
```
